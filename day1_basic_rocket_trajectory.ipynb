```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # gravity (m/s^2)
v0 = 100  # initial velocity (m/s)
angle = 45  # launch angle in degrees

# Velocity components
vx = v0 * np.cos(np.radians(angle))
vy = v0 * np.sin(np.radians(angle))

# Time simulation step-by-step
dt = 0.1
x_vals = []
y_vals = []
t = 0

while True:
    x = vx * t
    y = vy * t - 0.5 * g * t**2
    if y < 0:
        break
    x_vals.append(x)
    y_vals.append(y)
    t += dt

# Plot trajectory
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, color='blue')
plt.title('Rocket Trajectory Simulation')
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Vertical Height (m)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81        # Gravity (m/s^2)
drag = 0.1      # Air resistance coefficient
dt = 0.1        # Time step
t_total = 20    # Total time

# Initial conditions
v0 = 100
angle = 60
vx = v0 * np.cos(np.radians(angle))
vy = v0 * np.sin(np.radians(angle))
x, y = 0, 0

x_points, y_points = [], []
time = 0

while y >= 0:
    ax = -drag * vx
    ay = -g - drag * vy

    vx += ax * dt
    vy += ay * dt
    x += vx * dt
    y += vy * dt

    x_points.append(x)
    y_points.append(y)
    time += dt

plt.plot(x_points, y_points)
plt.title("Rocket Trajectory with Drag")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid(True)
plt.show()
```
